## Data Visualization

A line chart is created using Python that compares local temperatures (Delhi) with the global temperatures.


### Python3 code

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_excel('results.xlsx')
```

```python
df
```

```python
fig,axes = plt.subplots(figsize=(15,8))

sns.lineplot(x = 'year', y = 'city_MA', data = df, ax =axes,color='red',label='Local Temperature')

sns.lineplot(x = 'year', y = 'global_MA', data = df, ax =axes,color='blue',label='Global Temperature')
axes.legend()

plt.xlabel('Year')
plt.ylabel('Temperature (ºC)')
plt.title('Weather Trends')

axes.set_ylim([0, 40])
axes.set_xlim([1790,2015])
```

```python
fig.savefig('Weather Trends.png')
```
